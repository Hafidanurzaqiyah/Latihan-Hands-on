# Secara Visual

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
iris = pd.read_csv('./stars.csv')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris.plot(x='Temperature', y='Type')
```

```python
iris.plot(x='Temperature', y='Type', kind='scatter')
plt.xlabel('Temperature (T)')
plt.ylabel('percentation (?)')
```

```python
iris.plot(y='Temperature', kind='box')
plt.ylabel('Temperature (T)')
```

```python
iris.plot(y='Temperature', kind='hist')
plt.xlabel('Temperature (T)')
```

```python
iris.plot(y='Temperature', kind='hist', bins=30, range=(5000,40000), density=True)
plt.xlabel('Temperature (T)')
plt.show()
```

```python
iris.plot(y='Temperature', kind='hist', bins=30, range=(5000,40000), cumulative=True, density=True)
plt.xlabel('Temperature (T)')
plt.title('Cumulative distribution function (CDF)')
```

# Filtering

```python
iris.head()
```

```python
iris['Temperature'].describe()
```

```python
iris['Temperature'].value_counts()
```

```python
iris['Temperature'].unique()
```

```python
index_Clasificate1 = iris['Temperature'] >= 8000
index_Clasificate2= iris['Temperature'] <= 6000
```

```python
Clasificate1 = iris[index_Clasificate1]
Clasificate2 = iris[index_Clasificate2]
```

```python
Clasificate1 ['Temperature'].unique()
```

```python
Clasificate2['Temperature'].unique()
```

```python
Clasificate1.plot(kind='hist', bins=50, range=(5000,30000), alpha=0.5)
plt.title('Clasificate1 data set')
plt.xlabel('Temperature [T]')
```
